```python
import pandas as pd
```

```python
df = pd.DataFrame()
df.index.name = '#'

index_row = 1
x0 = 0.00005
mag = 100
for k in range(10):
    x1 = x0 * (1 + k)
    x2 = x1 * mag
    df.at[index_row, 'af_init'] = 0
    df.at[index_row, 'af_step'] = x1
    df.at[index_row, 'af_max'] = x2

    index_row += 1

df
```

```python
file_json = 'doe_trial_001.json'
df.to_json(file_json)
```
