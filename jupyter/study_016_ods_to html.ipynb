# Pandas で ODS ファイルを読み込む

Reference: https://pandas.pydata.org/docs/reference/api/pandas.read_excel.html

If path_or_buffer is an OpenDocument format (.odf, .ods, .odt), then odf will be used.

```python
import numpy as np
import pandas as pd
```

## ODF ファイルを指定

```python
#file_body = '../doe/report_af_7011_001'
#file_body = '../doe/report_af_7203_001'
file_body = '../doe/report_af_8306_001'
file_ods = '%s.ods' % file_body
file_html = '%s.html' % file_body
df = pd.read_excel(file_ods)
df
```

## データフレームを HTML に変換して出力

```python
row_max, col_max = df.shape
list_html = list()

list_html.append('<table class="simple">\n')

list_html.append('<thead>\n')
list_html.append('<tr>\n')
for colhead in df.columns:
    list_html.append('<th nowrap>%s</th>\n' % colhead)
list_html.append('</tr>\n')
list_html.append('</thead>\n')

list_html.append('<tbody>\n')
for r in range(row_max):
    list_html.append('<tr>\n')
    for c in range(col_max):
        v = df.iat[r, c]
        if np.isnan(v):
            list_html.append('<td nowrap style="text-align: right;"></td>\n')
        elif c == 0 or c > 3:
            list_html.append('<td nowrap style="text-align: right;">%d</td>\n' % int(v))
        else: 
            list_html.append('<td nowrap style="text-align: right;">%.5f</td>\n' % v)
    list_html.append('</tr>\n')
list_html.append('</tbody>\n')

list_html.append('</table>\n')

with open(file_html, mode='w') as f:
    f.writelines(list_html)
```
