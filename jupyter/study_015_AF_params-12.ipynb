```python
import pandas as pd

df = pd.DataFrame()
df.index.name = '#'

index_row = 1
x0 = 0.00001
mag = 100
for k in range(10):
    x1 = x0 * (k + 1)
    x2 = x1 * mag
    df.at[index_row, 'af_init'] = x1
    df.at[index_row, 'af_step'] = x1
    df.at[index_row, 'af_max'] = x2

    index_row += 1

df
```

```python
file_json = 'doe_trial_003.json'
df.to_json(file_json)
```
