```python
import pandas as pd
```

```python
df = pd.DataFrame()
df.index.name = '#'

for r in range(40):
    n = r + 1
    x1 = 0.00005 * n
    x2 = x1 * 10
    df.at[n, 'af_init'] = 0
    df.at[n, 'af_step'] = x1
    df.at[n, 'af_max'] = x2

df
```

```python
file_json = 'doe_af.json'
df.to_json(file_json)
```
