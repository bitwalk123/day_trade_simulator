```python
import pandas as pd
```

```python
df = pd.DataFrame()
df.index.name = '#'

index_row = 1
for k in [0.00001, 0.0001, 0.001, 0.01]:
    for r in [1, 2, 3, 5, 7.5]:
        x1 = k * r
        x2 = x1 * 10
        df.at[index_row, 'af_init'] = 0
        df.at[index_row, 'af_step'] = x1
        df.at[index_row, 'af_max'] = x2

        index_row += 1

df
```

```python
file_json = 'doe_af.json'
df.to_json(file_json)
```

```python
pd.read_json(file_json)
```

```python
file_html = 'doe_af.html'
df.to_html(file_html)
```
